### This dataset has 15.4+ MB of memory.
### Let's do some casts for reduce a memory used.


```python

df['Superfície'] = df['Superfície'].astype('category')
df['Jurisdição/Administração'] = df['Jurisdição/Administração'].astype('category')
df['Tipo de Trecho'] = df['Tipo de Trecho'].astype('category')
df['Tecnologia'] = df['Tecnologia'].astype('category')
df['Operadora'] = df['Operadora'].astype('category')
df['UF'] = df['UF'].astype('category')
df['Ano'] = df['Ano'].astype('int32')
df['SNV'] = df['SNV'].astype('category')
df['Rodovia'] = df['Rodovia'].astype('category')
df['Trecho'] = df['Trecho'].astype('category')
```
